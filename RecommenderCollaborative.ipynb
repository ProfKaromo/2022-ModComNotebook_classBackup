```python
# Colaborative - Get user reviews/ratings on a product/services
# Content  - This uses content to recommend...
```

```python
# Collaborative Filtering.
import pandas
movie_names = pandas.read_csv("https://modcom.co.ke/datasets/Movie_Id_Titles.csv")
movie_names.head(2)
```

```python
movie_details = pandas.read_csv("https://modcom.co.ke/datasets/file.csv")
movie_details.head(2)
```

```python
data = pandas.merge(movie_names, movie_details, on = 'item_id')
data.tail(5)
```

```python
# Lets see which movie was rated high/low on average
data.groupby('title')['rating'].mean().sort_values(ascending=False)
```

```python
# Let see which movie had more/less users ratings  - How many users rated?
data.groupby('title')['rating'].count().sort_values(ascending=False)
```

```python
# Combine average rating and number of ratings
ratings = pandas.DataFrame(data.groupby('title')['rating'].mean())
ratings['no_of_rating']  = data.groupby('title')['rating'].count()
ratings.sort_values('no_of_rating',ascending=False)
```

```python
# Pivot table 
pivot = data.pivot_table(index= 'user_id', columns= 'title', values= 'rating')
pivot
```

```python
# Correlation with Pivot
selected = pivot['Pulp Fiction (1994)']

# Know similar
similar  = pivot.corrwith(selected)

# Do a dataframe
similar_movies = pandas.DataFrame(similar, columns = ['Correlations'])
#similar_movies.sort_values('Correlations', ascending=False).head(20)

# How many people rated this movies
# Above correlation of 0.5 and number rating > 100
similar_movies = similar_movies.join(ratings['no_of_rating'])
similar_movies[similar_movies['no_of_rating'] > 200].sort_values('Correlations', ascending=False).head(20)
```
